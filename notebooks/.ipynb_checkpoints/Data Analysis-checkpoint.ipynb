```python
def show_views(volume, dir):
    D, H, W = volume.shape
    
    mid_d = D // 2
    mid_h = H // 2
    mid_w = W // 2
    
    axial = volume[mid_d]          # (H, W)
    coronal = volume[:, :, mid_w]  # (D, H)
    sagittal = volume[:, mid_h, :] # (D, W)

    plt.figure(figsize=(15, 5))
    
    plt.subplot(1, 3, 1)
    plt.title("Axial View")
    plt.imshow(axial, cmap="gray")
    plt.axis("off")

    plt.subplot(1, 3, 2)
    plt.title("Coronal View")
    plt.imshow(coronal, cmap="gray")
    plt.axis("off")

    plt.subplot(1, 3, 3)
    plt.title("Sagittal View")
    plt.imshow(sagittal, cmap="gray")
    plt.axis("off")

    plt.tight_layout()
    plt.savefig(results_dir + '/mri_slice_views_before.png', dpi=500)
    plt.show()
```

```python
results_dir = os.path.join(parent_directory,'results')
show_views(volume, results_dir)
```

# 📝 Data Observations and Preprocessing Decisions

## 1. T2-weighted MRI Volumes

**Example slice metadata:**

- **Image Origin:** (-107.93, -66.04, -74.88)  
- **Image Size:** (256, 256, 1)  
- **Image Spacing:** (0.664, 0.664, 1.0) mm  
- **Image Direction:** Identity with small numerical offsets  
- **Pixel Type:** 16-bit unsigned integer  

**Z-axis information:**

- **First 5 Z positions:** [-74.88, -73.38, -71.88, -70.38, -68.88]  
- **Z diffs:** [1.5, 1.5, 1.5, 1.5]  
- **Median Z spacing:** 1.5 mm  

**Observations:**

- Volume has **consistent 60 slices per patient**, each slice of size 256×256 pixels.  
- In-plane resolution (X/Y) = 0.664 mm, out-of-plane (Z) spacing = 1.5 mm → **anisotropic volume**.  
- Slices appear high-resolution in-plane, but Z spacing is coarser, motivating **super-resolution in the Z-axis**.

---

## 2. DWI / ADC / DCE Volumes

**Example slice metadata:**

- **Image Origin:** (-152.93, -88.29, -64.83)  
- **Image Size:** (160, 132, 1)  
- **Image Spacing:** (1.625, 1.625, 3.6) mm  
- **Pixel Type:** 16-bit unsigned integer  

**Observations:**

- Lower in-plane resolution (160×132 vs 256×256) → more coarse details.  
- Very thick slices in Z-axis (3.6 mm) → highly anisotropic.  
- Many slices might be missing or inconsistent across patients.  
- DWI, ADC, DCE represent **functional or derived maps**, not anatomical T2-weighted images.  
  - DWI / ADC highlights diffusion properties  
  - DCE captures perfusion over time  
  - These modalities are **not ideal for structural slice interpolation**, since intensities vary dynamically and may be inconsistent across slices.

---

## 3. Decision: Use only T2-weighted MRI

**Reasons:**

1. **Consistency:** T2-weighted MRI slices are uniform across patients (60 slices each, same in-plane resolution).  
2. **Anatomical relevance:** Middle-slice prediction / super-resolution focuses on **structural continuity**, which is best represented in T2-weighted images.  
3. **Feasible Z-axis interpolation:** With median 1.5 mm spacing, the model can reliably predict intermediate slices.  
4. **Avoid artifacts:** Functional volumes (DWI, ADC, DCE) have large spacing and variable intensities → predicting middle slices would be less meaningful and prone to errors.  

**Conclusion:**  
- For this project, we will **focus exclusively on T2-weighted MRI volumes** to generate intermediate slices and improve isotropic resolution in the Z-axis.
